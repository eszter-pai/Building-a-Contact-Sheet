```python
import PIL
from PIL import Image
from PIL import ImageEnhance
from PIL import ImageDraw
from PIL import ImageFont

# read image and convert to RGB
image=Image.open("readonly/msi_recruitment.gif")
image=image.convert('RGB')


channels = [0, 1, 2]
intensities = [0.1, 0.5, 0.9]

def newimage(c_index, i_index):
    
    # create the text area:
    font = ImageFont.truetype("readonly/fanwood-webfont.ttf", 50)
    text_area = Image.new('RGB', (image.width, 50))
    text_draw = ImageDraw.Draw(text_area)
    image_base=PIL.Image.new('RGB', (image.width, image.height+50))
    text_draw.text((0, 0), 'channel {} intensity {}'.format(channels[c_index], intensities[i_index]), font = font)
    
    # combine the text area and the image
    image_base.paste(text_area, (0, 450))
    image_base.paste(image, (0, 0))


    # alter RGB value of images:
    #800*450
    for x in range(0, 799):
        for y in range(0, 449+50):
            [r,g,b]=image_base.getpixel((x, y))
            if c_index == 0 and i_index == 0:
                image_base.putpixel((x, y), (int(r*0.1), g, b))
                        
            if c_index == 0 and i_index == 1:
                image_base.putpixel((x, y), (int(r*0.5), g, b))
         
            if c_index == 0 and i_index == 2:
                image_base.putpixel((x, y), (int(r*0.9), g, b))
              
            if c_index == 1 and i_index == 0:
                image_base.putpixel((x, y), (r, int(g*0.1), b))
                        
            if c_index == 1 and i_index == 1:
                image_base.putpixel((x, y), (r, int(g*0.5), b))
                    
            if c_index == 1 and i_index == 2:
                image_base.putpixel((x, y), (r, int(g*0.9), b))
                                            
            if c_index == 2 and i_index == 0:
                image_base.putpixel((x, y), (r, g, int(b*0.1)))
         
            if c_index == 2 and i_index == 1:
                image_base.putpixel((x, y), (r, g, int(b*0.5)))
        
            if c_index == 2 and i_index == 2:
                image_base.putpixel((x, y), (r, g, int(b*0.9)))
    return image_base

# build a list of 9 images which have different brightnesses

images=[]
for c in range(3):
    for i in range(3):
        images.append(newimage(c, i))

first_image=images[0]
contact_sheet=PIL.Image.new(first_image.mode, (first_image.width*3,first_image.height*3))
a=0
b=0

for img in images:
    # Lets paste the current image into the contact sheet
    contact_sheet.paste(img, (a, b) )
    # Now we update our X position. If it is going to be the width of the image, then we set it to 0
    # and update Y as well to point to the next "line" of the contact sheet.
    if a+first_image.width == contact_sheet.width:
        a=0
        b=b+first_image.height
    else:
        a=a+first_image.width

# resize and display the contact sheet
contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))
display(contact_sheet)
```
